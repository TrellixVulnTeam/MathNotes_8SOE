# ラグランジュの未定乗数法

### 概要
- 制約付き極値問題を解く際に用いられる有名なアルゴリズム
- ある関数f(x, y)を束縛条件g(x, y) = 0で最大化・最小化する解(x,y)は以下の方程式を解くことで得られる。
- 二変数以上の多変数でも適用できる。（が、簡単のため二変数で考える。）

$$ L(x, y) = f(x, y) + \lambda g(x, y) \tag{1} $$
$$ \frac{ \partial L }{ \partial x } = \frac{ \partial L }{ \partial y } = \frac{ \partial L }{ \partial \lambda } = 0 \tag{2}$$

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

## 1 : 等高線を考える

- まず、f(x, y), g(x,y)　は以下のような面を描く。

```python
def func1(x, y):
    return x**2 + y**2

def func2(x, y):
    return x**2 - 5*y

if __name__ == '__main__':
    x = np.arange(-3.0, 3.0, 0.1)
    y = np.arange(-3.0, 3.0, 0.1)
    X, Y = np.meshgrid(x, y)
    Z = func1(X, Y)
    Z2 = func2(X,Y)
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("f(x, y)")
    ax.plot_surface(X, Y, Z)
    ax.plot_surface(X, Y, Z2)
    plt.show()
```
